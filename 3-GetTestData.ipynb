```python
meanTemp = {}
date = list(data['date'])
testdate = list(data['date'])
temp = list(data['temperatureMean'])
for d,t in zip(date,temp):
    if d in testdate:
        meanTemp[d] = round(((t-32)/1.8),4)
```

```python
with open("data_temp.txt", 'w') as f:
    for key, value in meanTemp.items():
        f.write('%s   :   %s\n' % (key, value))
```
